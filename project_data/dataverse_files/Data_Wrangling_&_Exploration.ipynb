```python
from PIL import Image
import numpy as np
import os
import random as rd
```

```python
paths = ['HAM10000_images_part_1', 'HAM10000_images_part_2',
         'ISIC2018_Task3_Test_Images'] # 3 folders holding every image

png_list = []

for path in paths :
    # add folder image is in, '/', and png name to get pathway to all images
    png_list += ['./' + path + '/' + x for x in os.listdir(path)]

# test if all 11527 images are accounted for
print(len(png_list))



```

```python
# 1 random image from each folder
images = [png_list[rd.randint(0, 4999)], png_list[rd.randint(5000, 10013)], png_list[rd.randint(10014, 11526)]]


for image in images :
    # image of skin
    skin_img = Image.open(image)
    # array of skin image
    array_img = np.array(skin_img)
    
    print(image)
    skin_img.show()
    print(array_img)
    print('image shape : ', skin_img.size, '   image type : ', skin_img.mode, '  array shape : ', array_img.shape)
    print()


```

All 11527 images have file locations pointed to in png_list.  To show the files could be turned into numpy arrays for use in machine learning 3 png files each randomly chosen from 1 of 3 different directories with png files were used.  Their array values were shown with the (450, 600, 3) array size matching the 600 by 450 image size.  Also, the 3 in the array represented the 3 values in RBG format indicating the color of a pixel.
